```python
# Data pre-processing
# My first attempt at converting the training data into CountVector

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

from sklearn import preprocessing
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report


# Read the training.csv
raw_data = pd.read_csv('training.csv')
raw_test = pd.read_csv('test.csv')
df = pd.DataFrame(raw_data)
ds = pd.DataFrame(raw_test)

#Some data cleaning and preprocessing
#Replace all the commas in the strings with a 'space'
    #This is required as CountVectorizer splits words with spaces
df['article_words'] = df['article_words'].str.replace(",", " ")
ds['article_words'] = ds['article_words'].str.replace(",", " ")

#Define the input (x)
x = df['article_words']
#Define the labelled output (y)
y = df['topic']


x_test = ds['article_words']
y_test = ds['topic']

#Convert the Pandas dataframe to an array
x_array = x.values #Training set
x_array_test = x_test.values #Test set


#Create and fit the CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer()
x_train_BoW = count.fit_transform(x_array)
x_test_BoW = count.transform(x_array_test)
#Just some testing
print("X train BoW shape:")
print(x_train_BoW.shape)
print("X test BoW shape:")
print(x_test_BoW.shape)


#Create and fit the TFIDFVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
x_train_TFIDF = tfidf.fit_transform(x_array)
x_test_TFIDF = tfidf.transform(x_array_test)
print("X train TFIDF shape:")
print(x_train_TFIDF.shape)
print("X test TFIDF shape:")
print(x_test_TFIDF.shape)
# print(tfidf.vocabulary_) <- This is to check everything is nicely vectorized


#Encode the labelled data
label_enc = preprocessing.LabelEncoder()
#print(label_enc.classes_)
y_train = label_enc.fit_transform(y)
y_test = label_enc.transform(y_test)

print("Y train shape:")
print(y_train.shape)
print("Y test shape:")
print(y_test.shape)
#print(label_enc.classes_)


#Really quick SVM example on how to use the data the preprocessed Data. 

from sklearn import svm
SVM = svm.SVC(C=1.0, kernel='linear', degree=3, gamma='auto')
SVM.fit(x_train_TFIDF, y_train)
predictions_SVM = SVM.predict(x_test_TFIDF)
print("SVM Accuracy Score -> ",accuracy_score(predictions_SVM, y_test)*100)

```
